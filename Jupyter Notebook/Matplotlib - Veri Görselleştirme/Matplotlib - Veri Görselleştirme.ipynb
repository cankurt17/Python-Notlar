```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.arange(1,6)
```

```python
x
```

```python
y = np.arange(2,11,2)
```

```python
y
```

```python
plt.plot(x,y,"red")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,"blue")
plt.subplot(2,2,2)
plt.plot(y,x,"yellow")
plt.subplot(2,2,3)
plt.plot(x,y**2,"red")
plt.subplot(2,2,4)
plt.plot(x,y/2,"green")
plt.show()
```

```python
x = np.arange(1,6)
y = np.arange(2,11,2)

fig = plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.2,0.3])

axes1.plot(y,x)
axes1.set_xlabel("Outer X")
axes1.set_ylabel("Outer Y")
axes1.set_title("Outer Graph")

axes2.plot(x,y)
axes2.set_xlabel("Outer X")
axes2.set_ylabel("Outer Y")
axes2.set_title("Outer Graph")

plt.show()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1)

axes[0].plot(x,y)
axes[0].set_title("First Axes")
axes[1].plot(x,x**3)
axes[1].set_title("Second Axes")
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(5,3))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color="purple")
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,6))

axes[0].plot(x,y,color="red")
axes[0].set_title("Birinci")

axes[1].plot(x,x**2,color="purple")
axes[1].set_title("İkinci")

plt.tight_layout()
```

```python
fig = plt.figure(figsize=(6,4))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**0.5,color="red",label="karekök")
axes.plot(x,x**2,color="yellow",label="kare")
axes.plot(x,x**3,color="#4055c2",label="küp")
axes.legend()
```

```python
fig = plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x**2,color="red",lw=3,ls="--",marker="o",markersize=10,markerfacecolor="black")
axes.set_xlim(0,10)
axes.set_ylim(0,50)
```
